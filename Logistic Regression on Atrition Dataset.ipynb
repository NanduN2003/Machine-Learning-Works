```python
from  sklearn.preprocessing import LabelEncoder
Department_le=LabelEncoder()
Mar_le=LabelEncoder()
lab_df["Department"]=Department_le.fit_transform(lab_df["Department"])
```

```python
lab_df["MaritalStatus"]=Mar_le.fit_transform(lab_df["MaritalStatus"])
lab_df["Attrition"]=Department_le.fit_transform(lab_df["Attrition"])
lab_df["OverTime"]=Department_le.fit_transform(lab_df["OverTime"])
```

```python
objcol = lab_df.select_dtypes(["object"])
```

```python
objcol
```

```python
x = lab_df.drop(objcol,axis=1)
```

```python
x.drop(columns="Attrition",inplace= True)
y=lab_df["Attrition"]
y
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
li = LogisticRegression()
li.fit(X_train, Y_train)
```

```python
y_pred=li.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
a=mean_absolute_error(y_pred,Y_test)
```

```python
print("Accuracy=",1-a)
```

```python
from sklearn import metrics
print(metrics.classification_report(Y_test, y_pred, digits =3))
```

```python
b=metrics.confusion_matrix(Y_test, y_pred)
```

```python
sns.heatmap(b,annot= True)
```
